```python
# Data loading
file_path = "Resources/shopping_data.csv"
df_shopping = pd.read_csv(file_path, encoding="ISO-8859-1")
df_shopping.head(5)
```

```python
# What data is available?
# Columns
df_shopping.columns
```

```python
# What type of data is available?
# List dataframe data types
df_shopping.dtypes
```

```python
# What data is missing?
# Find all null values
for column in df_shopping.columns:
    print(f"Column {column} has {df_shopping[column].isnull().sum()} null values")
```

```python
# Drop null rows
df_shopping = df_shopping.dropna()
```

```python
# Confirm no more null rows:
for column in df_shopping.columns:
    print(f"Column {column} has {df_shopping[column].isnull().sum()} null values")
```

```python
# Find duplicate entries
print(f"Duplicate entries: {df_shopping.duplicated().sum()}")
```

```python
# Remove the CustomerID column - no useful info
df_shopping.drop(columns=["CustomerID"], inplace=True)
df_shopping.head()
```

```python
# Transform String column ("Card Member")
def change_string(member):
    if member == "Yes":
        return 1
    else:
        return 0

df_shopping["Card Member"] = df_shopping["Card Member"].apply(change_string)
df_shopping.head()
```

```python
# Transform "Anual Income"
# Scale for consistent number size between columns
df_shopping["Annual Income"] = df_shopping["Annual Income"] / 1000
df_shopping.head()
```

```python
# Replace spaces in column names with '_'
new_cols = []
for col in df_shopping.columns:
    new_cols.append(col.replace(' ', '_'))

# Remove Numbers from "Spending_Score"
new_cols[-1] = "Spending_Score"

df_shopping.columns = new_cols
df_shopping.head()
```

```python
# Store cleaned data as csv
# Saving cleaned data
file_path = "Resources/shopping_data_cleaned.csv"
df_shopping.to_csv(file_path, index=False)
```

```python
# Store as excel file for fun
df_shopping.to_excel("Resources/shopping_data_cleaned.xlsx")
```
